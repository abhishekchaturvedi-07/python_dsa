## BUBBLE SORT

```python
def bubble_sort(array):
    n = len(array)

    for i in range(n-1):
        for j in range(0, n-1):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
    return array
```

```python
import numpy as np
d = np.array([12,34,11,56,76,2,4,68,23,87])
```

```python
bubble_sort(d)
```
